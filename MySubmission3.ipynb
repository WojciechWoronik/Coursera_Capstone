First I'm importing required libraries:

```python
import pandas as pd
import numpy as np
```

Next, importing already created PostCode data file (from the first part of instructions) and geospatial CSV file:

```python
toronto_df = pd.read_csv('Toronto_df.csv')
del toronto_df['Unnamed: 0']
toronto_df.head()
```

```python
geo_df = pd.read_csv('Geospatial_Coordinates.csv')
geo_df.head()
```

Next, mapping based on Postal Code and adding longitude and latitude coordinates:

```python
geo_df = geo_df.set_index(['Postal Code'])
toronto_df = toronto_df.set_index(['Postcode'])

toronto_df.loc[:, 'Latitude'] = geo_df['Latitude']
toronto_df.loc[:, 'Longitude'] = geo_df['Longitude']

toronto_df.head(10)
```
